````python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker as mtick
import seaborn as sns

plt.rcParams['figure.dpi'] = 120
sns.set_palette('husl')
print('Libraries loaded ✓')
````

## 1. Loading Data

Obesity prevalence (% adult population) by country 2000–2024. Source: WHO / Mendeley Data k8ksc5zxyr

````python
data = {
    'Country': ['Mexico','Brazil','Argentina','Colombia','Chile','Peru','Venezuela','Ecuador','Bolivia','Paraguay'],
    '2000': [24.2, 18.3, 22.1, 18.7, 24.0, 15.4, 24.0, 17.1, 19.2, 18.9],
    '2010': [30.2, 22.1, 26.8, 22.4, 28.6, 18.3, 28.9, 21.2, 23.8, 23.6],
    '2020': [34.5, 25.9, 30.1, 26.7, 33.6, 21.4, 33.2, 24.9, 27.3, 28.1],
    '2024': [36.1, 27.4, 31.8, 28.1, 35.8, 22.7, 35.1, 26.4, 29.1, 30.2]
}
df = pd.DataFrame(data).set_index('Country')
df
````

````python
fig, axes = plt.subplots(1, 2, figsize=(16, 6))

# Chart 1: Heatmap
ax1 = axes[0]
sns.heatmap(df, ax=ax1, cmap='YlOrRd', annot=True, fmt='.1f', linewidths=0.5,
            cbar_kws={'label': 'Obesity Prevalence (%)'})
ax1.set_title('Obesity Prevalence Heatmap\nLatin America (% adult population)', fontsize=12, fontweight='bold')

# Chart 2: 2000 vs 2024 comparison
ax2 = axes[1]
x = range(len(df))
width = 0.35
ax2.bar([i-width/2 for i in x], df['2000'], width, label='2000', color='#3498db', alpha=0.8)
ax2.bar([i+width/2 for i in x], df['2024'], width, label='2024', color='#e74c3c', alpha=0.8)
ax2.set_title('Obesity Comparison: 2000 vs 2024', fontsize=12, fontweight='bold')
ax2.set_xticks(list(x))
ax2.set_xticklabels(df.index, rotation=45, ha='right')
ax2.set_ylabel('Prevalence (%)')
ax2.legend()
ax2.grid(True, alpha=0.3, axis='y')

plt.tight_layout()
plt.savefig('obesity_latam_trends.png', dpi=150, bbox_inches='tight')
plt.show()
print('Chart saved: obesity_latam_trends.png')
````

````python
print('=== KEY STATISTICS ===')
print(f'Highest obesity 2024: {df["2024"].idxmax()} ({df["2024"].max():.1f}%)')
print(f'Lowest obesity 2024: {df["2024"].idxmin()} ({df["2024"].min():.1f}%)')
print(f'Regional average 2024: {df["2024"].mean():.1f}%')
print(f'Average increase 2000-2024: +{(df["2024"]-df["2000"]).mean():.1f} percentage points')
````
